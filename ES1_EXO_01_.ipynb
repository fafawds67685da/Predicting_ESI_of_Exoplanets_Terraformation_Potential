```python
import matplotlib.pyplot as m
import numpy as n
import pandas as p

det=n.random.randint(1,3,100)
mass=n.random.randint(1,100,100)
radius=n.random.randint(1,100,100)
radiation=n.random.randint(1,100,100)
surf_temp=n.random.randint(1,100,100)
orbital_period=n.random.randint(1,100,100)
distance=n.random.randint(1,100,100)
density=n.random.randint(1,100,100)
gravity=n.random.randint(1,100,100)

esi=n.random.randint(1,100,100)

def scale(x):
  return (x-n.min(x))/(n.max(x)-n.min(x))

det1=scale(det)
mass1=scale(mass)
radius1=scale(radius)
radiation1=scale(radiation)
surf_temp1=scale(surf_temp)
orbital_period1=scale(orbital_period)
distance1=scale(distance)
density1=scale(density)
gravity1=scale(gravity)
esi1=scale(esi)


print(det1)
print(mass1)
print(radius1)
print(radiation1)
print(surf_temp1)
print(orbital_period1)
print(distance1)
print(density1)
print(gravity1)
print(esi1)
```

```python
w1=0
w2=0
w3=0
w4=0
w5=0
w6=0
w7=0
w8=0
w9=0

lr=0.0001
b=0
```

```python
def predict(x1,x2,x3,x4,x5,x6,x7,x8,x9,w1,w2,w3,w4,w5,w6,w7,w8,w9,b):
  z1=x1*w1
  z2=x2*w2
  z3=x3*w3
  z4=x4*w4
  z5=x5*w5
  z6=x6*w6
  z7=x7*w7
  z8=x8*w8
  z9=x9*w9
  return z1+z2+z3+z4+z5+z6+z7+z8+z9+ b

def cost_function(det1,mass1,radius1,radiation1,surf_temp1,orbital_period1,distance1,density1,gravity1,esi1,w1,w2,w3,w4,w5,w6,w7,w8,w9,b):
  y=predict(det1,mass1,radius1,radiation1,surf_temp1,orbital_period1,distance1,density1,gravity1,w1,w2,w3,w4,w5,w6,w7,w8,w9,b)
  mse=n.mean((y-esi1)**2)
  return mse
```

```python
print(cost_function(det1,mass1,radius1,radiation1,surf_temp1,orbital_period1,distance1,density1,gravity1,esi1,w1,w2,w3,w4,w5,w6,w7,w8,w9,b))
```

```python
def update(det1,mass1,radius1,radiation1,surf_temp1,orbital_period1,distance1,density1,gravity1,esi1,w1,w2,w3,w4,w5,w6,w7,w8,w9,b,lr):
  dLdw=n.mean(-2*det1*(esi1-predict(det1,mass1,radius1,radiation1,surf_temp1,orbital_period1,distance1,density1,gravity1,w1,w2,w3,w4,w5,w6,w7,w8,w9,b)))
  dLdw2=n.mean(-2*mass1*(esi1-predict(det1,mass1,radius1,radiation1,surf_temp1,orbital_period1,distance1,density1,gravity1,w1,w2,w3,w4,w5,w6,w7,w8,w9,b)))
  dLdw3=n.mean(-2*radius1*(esi1-predict(det1,mass1,radius1,radiation1,surf_temp1,orbital_period1,distance1,density1,gravity1,w1,w2,w3,w4,w5,w6,w7,w8,w9,b)))

  dLdw4=n.mean(-2*radiation1*(esi1-predict(det1,mass1,radius1,radiation1,surf_temp1,orbital_period1,distance1,density1,gravity1,w1,w2,w3,w4,w5,w6,w7,w8,w9,b)))
  dLdw5=n.mean(-2*surf_temp1*(esi1-predict(det1,mass1,radius1,radiation1,surf_temp1,orbital_period1,distance1,density1,gravity1,w1,w2,w3,w4,w5,w6,w7,w8,w9,b)))
  dLdw6=n.mean(-2*orbital_period1*(esi1-predict(det1,mass1,radius1,radiation1,surf_temp1,orbital_period1,distance1,density1,gravity1,w1,w2,w3,w4,w5,w6,w7,w8,w9,b)))

  dLdw7=n.mean(-2*distance1*(esi1-predict(det1,mass1,radius1,radiation1,surf_temp1,orbital_period1,distance1,density1,gravity1,w1,w2,w3,w4,w5,w6,w7,w8,w9,b)))
  dLdw8=n.mean(-2*density1*(esi1-predict(det1,mass1,radius1,radiation1,surf_temp1,orbital_period1,distance1,density1,gravity1,w1,w2,w3,w4,w5,w6,w7,w8,w9,b)))
  dLdw9=n.mean(-2*gravity1*(esi1-predict(det1,mass1,radius1,radiation1,surf_temp1,orbital_period1,distance1,density1,gravity1,w1,w2,w3,w4,w5,w6,w7,w8,w9,b)))

  dLdb=n.mean(-2*(esi1-predict(det1,mass1,radius1,radiation1,surf_temp1,orbital_period1,distance1,density1,gravity1,w1,w2,w3,w4,w5,w6,w7,w8,w9,b)))

  w1=w1-lr*dLdw
  w2=w2-lr*dLdw2
  w3=w3-lr*dLdw3

  w4=w4-lr*dLdw4
  w5=w5-lr*dLdw5
  w6=w6-lr*dLdw6

  w7=w7-lr*dLdw7
  w8=w8-lr*dLdw8
  w9=w9-lr*dLdw9

  b=b-lr*dLdb

  return w1,w2,w3,w4,w5,w6,w7,w8,w9,b
```

```python
print(update(det1,mass1,radius1,radiation1,surf_temp1,orbital_period1,distance1,density1,gravity1,esi1,w1,w2,w3,w4,w5,w6,w7,w8,w9,b,lr))
```

```python
def train(det1,mass1,radius1,radiation1,surf_temp1,orbital_period1,distance1,density1,gravity1,esi,w1,w2,w3,w4,w5,w6,w7,w8,w9,b,lr,tol=1e-8,n=100,verbose=False):
  weights1=[]
  weights2=[]
  weights3=[]

  weights4=[]
  weights5=[]
  weights6=[]

  weights7=[]
  weights8=[]
  weights9=[]

  biases=[]
  costs=[]

  ct=1
  while True:
    cost=cost_function(det1,mass1,radius1,radiation1,surf_temp1,orbital_period1,distance1,density1,gravity1,esi1,w1,w2,w3,w4,w5,w6,w7,w8,w9,b)
    costs.append(cost)

    if len(costs)>1 and abs(costs[-1]-costs[-2])< tol or ct==200000 :
      break
    w1,w2,w3,w4,w5,w6,w7,w8,w9,b=update(det1,mass1,radius1,radiation1,surf_temp1,orbital_period1,distance1,density1,gravity,esi,w1,w2,w3,w4,w5,w6,w7,w8,w9,b,lr)

    weights1.append(w1)
    weights2.append(w2)
    weights3.append(w3)

    weights4.append(w4)
    weights5.append(w5)
    weights6.append(w6)

    weights7.append(w7)
    weights8.append(w8)
    weights9.append(w9)

    biases.append(b)

    if verbose and ct % n==0:
      print("After ",ct," :- W1= ",w1," :- W2= ",w2," :- w3= ",w3," b= ",b," costs= ",cost)
    ct+=1

  return weights1,weights2,weights3,weights4,weights5,weights6,weights7,weights8,weights9,biases,costs


```

```python
w1,w2,w3,w4,w5,w6,w7,w8,w9,b,costs=train(det1,mass1,radius1,radiation1,surf_temp1,orbital_period1,distance1,density1,gravity1,esi1,w1,w2,w3,w4,w5,w6,w7,w8,w9,b,lr,tol=1e-13,n=100,verbose=True)
```
